```python
for col in df.columns:
  if df[col].dtype=='object':# ustun object bo'lsa, mode qo'llaniladi
    df[col].fillna(df[col].mode()[0],inplace=True)
  else:
    df[col].fillna(df[col].mean(),inplace=True) #aks holda mean qo'llaniladi
```

mutual scoreni hisoblash uchun target qiymatni input qiymamtni topish

```python
 x=df.drop(columns=['House_Price'])
```

```python
x
```

```python
y=df['House_Price']
```

```python
y
```

```python
from sklearn.feature_selection import mutual_info_regression # sklearn kutubxoansidan xususiyatlarni tanlash qismidan mutual info regression funksiya tanlandi
```

```python
mi_score=mutual_info_regression(x,y,discrete_features='auto') # funksiya orqali x, y qiymatga ajratib olindi, discrete_features='auto' qiymatlarni avtomatik tarzda dikrit yoki continuos qiymatlarga ajratish, agar shunday avtomatik ajratilmasa, bu ko'p vaqt hamda quvvat oladi
```

```python
mi_score # natija
```

```python
mi_df=pd.DataFrame({'Feature':x.columns,'Mutual information Score':mi_score}) #dataframe aniq shaklda chiqarish uchun qo'llaniladi
```

```python
mi_df=mi_df.sort_values(by='Mutual information Score',ascending=False) #dataframe ustunlarining tartibi(order)=> ascending=False bo'lsa yuqoridan pastga, ascending=True bo'lsa pastdan yuqoriga tarzda qiymatlarni chiqaradi
```
